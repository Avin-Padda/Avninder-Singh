```python
df['age'] = df['age'].fillna(df['age'].mean())
```

```python
X = df.drop(['survived', 'parch', 'who', 'adult_male', 'deck', 'embark_town', 'alive', 'alone'], axis = 1)
y = df['survived']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
from sklearn.preprocessing import LabelEncoder

# Create a LabelEncoder for each column
le_sex = LabelEncoder()
le_embarked = LabelEncoder()
le_class = LabelEncoder()

# Fit and transform each column in the training set
X_train['sex'] = le_sex.fit_transform(X_train['sex'])
X_train['embarked'] = le_embarked.fit_transform(X_train['embarked'])
X_train['class'] = le_class.fit_transform(X_train['class'])

# Only transform (not fit) on the test set
X_test['sex'] = le_sex.transform(X_test['sex'])
X_test['embarked'] = le_embarked.transform(X_test['embarked'])
X_test['class'] = le_class.transform(X_test['class'])
```
